```python
import sys
import os

# Agregamos la carpeta src al path
src_path = os.path.abspath('src')
if src_path not in sys.path:
    sys.path.append(src_path)
```

```python
from generation.instances import generate_instances
from generation.data import generate_data
from generation.adapters import EnrichedStackMatrix5DAdapter, DefaultMovesAdapter

# 1. Define tus escenarios de prueba (H, S, N, cantidad)
# Puedes ajustar las cantidades según el tiempo de cómputo que tengas en Colab
configuraciones = [
    # (H, S, N, amount)
    (5, 4, 15, 50000),   # Tableros pequeños (fáciles)
    (7, 5, 25, 100000),  # Tu benchmark actual (intermedios)
    (10, 6, 45, 50000),  # Tableros grandes y altos (difíciles)
    (6, 7, 30, 50000)    # Tableros anchos pero bajos
]

r_desorden = 50
max_steps = 50
seed = 42

print("🚀 Iniciando pipeline de generación multi-tamaño...\n")

for H, S, N, amount in configuraciones:
    filename = f"E{S}-{N}-H{H}_V8"
    print(f"--- Procesando Configuración: {S} Stacks, {N} Contenedores, Altura {H} ---")

    # 1. Generar instancias brutas
    print(f"Generando {amount} instancias brutas...")
    generate_instances(filename, H, S, N, amount, r_desorden, seed)

    # 2. Extraer características 5D y resolver
    print("Resolviendo y empaquetando datos 5D...")
    layout_adapter = EnrichedStackMatrix5DAdapter()
    moves_adapter = DefaultMovesAdapter()

    generate_data(filename, H, max_steps, layout_adapter, moves_adapter)
    print(f"✅ Finalizado: {filename}.data guardado en la carpeta 'data/'.\n")

print("🎉 ¡Toda la generación multi-tamaño ha terminado!")
```
