```python

import sys
import os

# Agrega la carpeta 'src' al path de python
sys.path.append(os.path.abspath("../src"))
from preprocessing.dataset import load_dataset


dataset_s3 = load_dataset("E4-15-H5_V8.data")
dataset_s4  = load_dataset("E5-25-H7_V8.data")
dataset_s5   = load_dataset("E6-45-H10_V8.data")
dataset_s6  = load_dataset("E7-30-H6_V8.data")
```

```python
from torch.utils.data import ConcatDataset

data = ConcatDataset([dataset_s3, dataset_s4, dataset_s5, dataset_s6])

print(f"Dataset combinado con un total de {len(data)} instancias.")
```

```python
from models.cpmp_transformer_v9 import CPMPTransformer

H = 12
C_dim = 2
X_dim = 5
d_model = 64
nhead = 4
num_layers = 4
ff_dim_multiplier = 2
dropout = 0.3

model = CPMPTransformer(
    H=H,
    C_dim=C_dim,
    X_dim=X_dim,
    d_model=d_model,
    nhead=nhead,
    num_layers=num_layers,
    ff_dim_multiplier=ff_dim_multiplier,
    dropout=dropout
)
```

```python
from training.training import train, Accuracy

epochs = 10
train_size = int(len(data) * 0.8) # 80% para entrenar, calculado dinámicamente
test_size = len(data) - train_size # 20% para test
batch_size = 128
learning_rate = 4e-4
weight_decay = 1e-3
patience = 10
metrics = [Accuracy()]
seed = 42

model = train(model, epochs, data, train_size, test_size, batch_size, learning_rate, weight_decay, patience, metrics, seed)
```
