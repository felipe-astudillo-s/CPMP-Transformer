```python
import sys
import os

base = r"D:\Proyectos\CPMP_TRANSFORMER\CPMP-Framework\CPMP-Framework"
sys.path.insert(0, os.path.join(base, "src"))
sys.path.insert(0, base)  # para settings.py

from preprocessing.dataset import load_dataset
```

```python

dataset_pequeno = load_dataset("V9_DEFINITIVO_S4_H12_30k.data")
dataset_normal  = load_dataset("V9_DEFINITIVO_S5_H12_70k.data")
dataset_ancho   = load_dataset("V9_DEFINITIVO_S6_H12_70k.data")
dataset_grande  = load_dataset("V9_DEFINITIVO_S7_H12_50k.data")
dataset_grandee  = load_dataset("V9_DEFINITIVO_S8_H12_30k.data")
```

```python
from torch.utils.data import ConcatDataset

data = ConcatDataset([dataset_pequeno, dataset_normal, dataset_ancho, dataset_grande, dataset_grandee])

print(f"Dataset combinado con un total de {len(data)} instancias.")
```

```python
from models.cpmp_transformer_v9 import CPMPTransformer

H = 12
C_dim = 2
X_dim = 5
d_model = 64
nhead = 4
num_layers = 4
ff_dim_multiplier = 2
dropout = 0.3

model = CPMPTransformer(
    H=H,
    C_dim=C_dim,
    X_dim=X_dim,
    d_model=d_model,
    nhead=nhead,
    num_layers=num_layers,
    ff_dim_multiplier=ff_dim_multiplier,
    dropout=dropout
)
```

```python
from training.training import train, Accuracy

epochs = 30
train_size = int(len(data) * 0.8) # 80% para entrenar, calculado dinámicamente
test_size = len(data) - train_size # 20% para test
batch_size = 128
learning_rate = 4e-4
weight_decay = 1e-3
patience = 10
metrics = [Accuracy()]
seed = 42

model = train(model, epochs, data, train_size, test_size, batch_size, learning_rate, weight_decay, patience, metrics, seed)
```
