```python
import sys
import os

sys.path.append(os.path.abspath("../src"))
```

## 1. Cargar modelo

```python
from models.cpmp_transformer_v9 import CPMPTransformer
from training.training import load_model

model_name = "v9"
model = load_model(CPMPTransformer, model_name)
print(f"Modelo '{model_name}' cargado correctamente.")
```

## 2. Configuración del benchmark

```python
from solvers.model import ModelSolver
from solvers.benchmark import benchmark_cvs, print_benchmark_results, save_results_csv

solver = ModelSolver(model)

# max_steps_fn(S, H) -> número máximo de pasos para instancias de tamaño (S, H)
# Por defecto: max(50, S * H * 4)
# Puedes personalizar esta función según tus necesidades.
max_steps_fn = lambda s, h: max(50, s * h * 4)
```

## 3. Ejecutar benchmark sobre todas las instancias CVS

```python
print("Ejecutando benchmark en instancias CVS...\n")
results = benchmark_cvs(solver, max_steps_fn=max_steps_fn)
```

## 4. Resultados

```python
print_benchmark_results(results)
```

## 5. Exportar resultados a CSV (opcional)

```python
save_results_csv(results, "../results/benchmark_v9.csv")
```
