```python

import sys
import os

# Agrega la carpeta 'src' al path de python
sys.path.append(os.path.abspath("../src"))

from preprocessing.dataset import load_dataset

filename = "E5-25.data"
data = load_dataset(filename)
```

```python
from models.cpmp_transformer_v8 import CPMPTransformer

H = 7
C_dim = 3
X_dim = 5
d_model = 64
nhead = 4
num_layers = 4
ff_dim_multiplier = 2
dropout = 0.3

model = CPMPTransformer(
    H=H,
    C_dim=C_dim,
    X_dim=X_dim,
    d_model=d_model,
    nhead=nhead,
    num_layers=num_layers,
    ff_dim_multiplier=ff_dim_multiplier,
    dropout=dropout
)
```

```python
from training.training import train, Accuracy

epochs = 10
train_size = 160000 
test_size = 40000
batch_size = 128
learning_rate = 4e-4
weight_decay = 1e-3
patience = 10
metrics = [Accuracy()]
seed = 42

model = train(model, epochs, data, train_size, test_size, batch_size, learning_rate, weight_decay, patience, metrics, seed)
```
